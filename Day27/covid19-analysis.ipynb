# Exporting the data

```python
import pandas as pd
```

```python
df=pd.read_csv('Covid_Confirmed.csv')
```

```python
df1=pd.read_csv('Covid_Death.csv')
```

```python
df2=pd.read_csv('Covid_recovered.csv')
```

```python
df2.head()
```

```python
import matplotlib.pyplot as plt
```

```python
df=df.drop(["Province/State","Lat","Long"],axis=1)
```

```python
df.head()
```

```python
df=df.groupby(df["Country/Region"]).aggregate('sum')
```

```python
df2=df2.drop(["Province/State","Lat","Long"],axis=1)
df1=df1.drop(["Province/State","Lat","Long"],axis=1)
```

```python
df1=df1.groupby(df1["Country/Region"]).aggregate('sum')
df2=df2.groupby(df2["Country/Region"]).aggregate('sum')
```

```python
print(df2)
```

```python
df=df.T
df1=df1.T
df2=df2.T
```

```python
df.head()
```

```python
df1.head()
```

```python
df2.head()
```

```python
new_cases=df.copy()
for day in range(1, len(df)):
    new_cases.iloc[day]=df.iloc[day]-df.iloc[day-1]
print(new_cases.tail(10))
```

```python
growth_rate=df.copy()
for day in range(1,len(df)):
    growth_rate.iloc[day]= (new_cases.iloc[day]/df.iloc[day-1]) * 100
growth_rate
```

```python
active_cases=df.copy()
for day in range(1,len(df)):
    active_cases.iloc[day]= df.iloc[day]= df2.iloc[day]
active_cases
```

```python
overall_growth_rate =df.copy()

for day in range(1,len(df)):
    overall_growth_rate.iloc[day]=((active_cases.iloc[day] - active_cases.iloc[day-1]) /active_cases.iloc[day-1])*100
```

```python
overall_growth_rate['India'].tail(10)
```

```python
death_rate = df.copy()
for day in range(1,len(df)):
    death_rate.iloc[day]=(df1.iloc[day]/df.iloc[day])*100
```

```python
death_rate["India"].tail(10)
```

```python
hospitalization_rate_estimate = 0.05
hostpitalization_needed=df.copy()
for day in range(1,len(df)):
    hostpitalization_needed.iloc[day]= active_cases.iloc[day] * hospitalization_rate_estimate

hostpitalization_needed["India"].tail()  
        
```

# Visualization

## Confirmed Cases

```python
countries=["India","Russia","US","United Kingdom","Brazil","Canada","Italy"]
ax=plt.subplot()
ax.set_facecolor('black')
ax.figure.set_facecolor('#121212')
ax.tick_params(axis='x',colors='white')
ax.tick_params(axis='y',colors='white')
ax.set_title('Covid-19 Total Confirmed Cases by cases',color='white')
for country in countries:
    df[country][100:].plot(label=country,figsize=(10,8))
plt.legend(loc='upper left')
plt.show()
```

## Growth rate

```python
coun=["India"]
ax=plt.subplot()
ax.set_facecolor('black')
ax.figure.set_facecolor('#121212')
ax.tick_params(axis='x',colors='white')
ax.tick_params(axis='y',colors='white')
ax.set_title('Covid-19 - Confirmed Cases growth rate (Coun)',color='white')
for coun2 in coun:
    growth_rate[country][200:].plot.bar(figsize=(100,80))
plt.show()

```

## Confimed Deaths

```python
coun=["India","US","Russia","Italy"]
for country in coun:
    ax=plt.subplot()
    ax.set_facecolor('black')
    ax.figure.set_facecolor('#121212')
    ax.tick_params(axis='x',colors='white')
    ax.tick_params(axis='y',colors='white')
    ax.set_title(f'Covid-19 - Confirmed Deaths in {country}',color='white')

    df1[country][200:].plot.bar(label=country,figsize=(15,12))
    plt.show()
```

## Death rate

```python
coun=["India","US","Russia","Italy","Germany"]
for country in coun:
    ax=plt.subplot()
    ax.set_facecolor('black')
    ax.figure.set_facecolor('#121212')
    ax.tick_params(axis='x',colors='white')
    ax.tick_params(axis='y',colors='white')
    ax.set_title(f'Covid-19 - Death rate in {country}',color='white')

    death_rate[country][200:].plot.bar(label=country,figsize=(15,12))
    plt.show()
```

```python
simulated_growth_rate=0.01
dates= pd.date_range(start='10/27/2020',periods=20,freq='D')
dates= pd.Series(dates)
dates= dates.dt.strftime("%m/%d/%y")
simulated= df.copy()
simulated=simulated.append(pd.DataFrame(index=dates))

for day in range(len(df), len(df)+20):
    simulated.iloc[day]= simulated.iloc[day-1]* (1+simulated_growth_rate)
simulated["Canada"]
```
