# Rainfall in India Data Analysis

```python
import pandas as pd 
import seaborn as sns
import numpy as np 
import matplotlib.pyplot as plt
from subprocess import check_output
```

```python
df = pd.read_csv('rainfall in india 1901-2015.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.drop(["YEAR"],axis=1).hist(figsize=(12,12),color='red')
```

```python
df.groupby("YEAR").sum()['ANNUAL'].plot(figsize=(12,8),color='red');
```

```python
df.columns
```

```python
df[['YEAR', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("YEAR").sum().plot(figsize=(13,8));
```

```python
df[['YEAR','Jan-Feb', 'Mar-May',
       'Jun-Sep', 'Oct-Dec']].groupby("YEAR").sum().plot(figsize=(13,8))
```

```python
df[['SUBDIVISION', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("SUBDIVISION").mean().plot.barh(stacked=True,figsize=(13,8))
```

```python
df[['SUBDIVISION', 'Jan-Feb', 'Mar-May',
       'Jun-Sep', 'Oct-Dec']].groupby("SUBDIVISION").sum().plot.barh(stacked=True,figsize=(16,8))
```

```python
df[['Jan-Feb', 'Mar-May',
       'Jun-Sep', 'Oct-Dec']].plot(kind="kde",figsize=(10,8))
```

```python
df[['SUBDIVISION', 'JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL',
       'AUG', 'SEP', 'OCT', 'NOV', 'DEC']].groupby("SUBDIVISION").max().plot.barh(figsize=(16,8))
```
